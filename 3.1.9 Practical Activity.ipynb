# 3.1.9 Practical Activty

```python
# Import Pandas.
import pandas as pd
```

```python
# Load the Excel and CSV files
movies = pd.read_excel('movies_merge.xlsx')
ott = pd.read_csv('ott_merge.csv')
```

```python
# Merge the two DataFrames.
mov_ott = pd.merge(movies, ott, how='left', on = 'ID')
```

```python
# Check DataFrames merged correctly
print(mov_ott.columns)
print(mov_ott.shape)
```

```python
mov_ott
```

# Q1: How many films from each year (released from 2012 to the present) were watched on Netflix?

```python
# Determine the number of movies watched on Netflix since 2012.
mo_gpby = mov_ott.groupby(['Year'])['Netflix'].agg('sum').reset_index()
mo_gpby[mo_gpby['Year'] >= 2012]
```

```python
# Practise again
```

```python
# mov_grp is name for new groupby DF
```

```python
mo_grp = mov_ott.groupby(['Year'])['Netflix'].agg('sum').reset_index()
```

```python
# Prune it down to only films released from 2012 and onwards
```

```python
mo_grp[mo_grp['Year'] >= 2012]
```

# Q2 Mean length of films each year

```python
# Create a new DF to hold the groupby data
```

```python
mo_runtim = mov_ott.groupby(['Year'])['Runtime'].agg('mean').reset_index()
```

```python
# Filter on films released from 2012 and onwards
```

```python
mo_runtim[mo_runtim['Year'] >=2012]
```

# Q3 Find min() and max() Rotten Tomatoe values 2012 onwards

```python
# Create new DF called mo_tom to hold groupby data
```

```python
mo_tom = mov_ott.groupby(['Year'])['Rotten Tomatoes'].agg(['max', 'min']).reset_index()
```

```python
mo_tom[mo_tom['Year'] >=2012]
```
